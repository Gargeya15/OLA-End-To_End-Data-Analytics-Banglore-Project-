```python
print("\nUNIQUE VALUES (Categorical):")
for col in df.select_dtypes(include='object').columns:
    print(f"\n{col}:", df[col].unique()[:5])  
```

```python
print(df.describe())
```

```python
# Check spaces
for col in df.select_dtypes(include='object').columns:
    if df[col].str.startswith(" ").any() or df[col].str.endswith(" ").any():
        print(f"Extra spaces in: {col}")
```

```python
# Check 'Error' values
for col in df.columns:
    if df[col].astype(str).str.contains("Error").any():
        print(f"Invalid 'Error' values in: {col}")
```

```python
# 1. Remove duplicates
df = df.drop_duplicates()
```

```python
# 2. Clean text (spaces + case)
for col in df.select_dtypes(include='object').columns:
    df[col] = df[col].str.strip().str.title()
```

```python
# 3. Replace 'Error' and convert numeric
for col in df.columns:
    if df[col].astype(str).str.contains("Error").any():
        df[col] = pd.to_numeric(df[col], errors='coerce')
```

```python
# 4. Handle missing values
for col in df.select_dtypes(include=np.number).columns:
    df[col] = df[col].fillna(df[col].median())
```

```python
for col in df.select_dtypes(include='object').columns:
    df[col] = df[col].fillna(df[col].mode()[0])
```

```python
# 5. Handle outliers (capping)
for col in df.select_dtypes(include=np.number).columns:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    df[col] = np.where(df[col] < lower, lower, df[col])
    df[col] = np.where(df[col] > upper, upper, df[col])
```

```python
print("Shape:", df.shape)
```

```python
print("\nMissing Values:\n", df.isnull().sum())
```
